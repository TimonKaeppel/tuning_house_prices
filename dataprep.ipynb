```python
import pandas as pd

results_pairwise_df = pd.read_csv('resources/results_pairwise.csv')
results_pairwise_df.head()
```

```python
# split df into dfs

df_num_trees_mtry = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'num.trees')
    &
    (results_pairwise_df['Parameter2'] == 'mtry')
    ]

df_num_trees_sample_fraction = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'num.trees')
    &
    (results_pairwise_df['Parameter2'] == 'sample.fraction')
    ]

df_num_trees_min_node_size = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'num.trees')
    &
    (results_pairwise_df['Parameter2'] == 'min.node.size')
    ]

df_num_trees_replace = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'num.trees')
    &
    (results_pairwise_df['Parameter2'] == 'replace')
    ]

df_mtry_sample_fraction = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'mtry')
    &
    (results_pairwise_df['Parameter2'] == 'sample.fraction')
    ]

df_mtry_min_node_size = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'mtry')
    &
    (results_pairwise_df['Parameter2'] == 'min.node.size')
    ]

df_mtry_replace = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'mtry')
    &
    (results_pairwise_df['Parameter2'] == 'replace')
    ]

df_sample_fraction_min_node_size = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'sample.fraction')
    &
    (results_pairwise_df['Parameter2'] == 'min.node.size')
    ]

df_sample_fraction_replace = results_pairwise_df[
    (results_pairwise_df['Parameter1'] == 'sample.fraction')
    &
    (results_pairwise_df['Parameter2'] == 'replace')
    ]

dfs = [df_num_trees_mtry, df_num_trees_sample_fraction, df_num_trees_min_node_size, df_num_trees_replace,
       df_mtry_sample_fraction, df_mtry_min_node_size, df_mtry_replace,
       df_sample_fraction_min_node_size, df_sample_fraction_replace]

for df in dfs:
    param_1_name = df['Parameter1'].values[0]
    param_2_name = df['Parameter2'].values[0]
    df = df.rename(columns={'Value1': param_1_name, 'Value2': param_2_name})
    df = df.drop(columns=['Parameter1', 'Parameter2'])
    df_name = f'{param_1_name}_{param_2_name}'

    print(df.head())
    print('\n')

    with open(f'resources/results_pairwise/{df_name}.csv', 'w') as f:
        f.write(df.to_csv(index=False))
```
